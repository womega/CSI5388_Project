```python
getattr(tqdm, '_instances', {}).clear()
#proc_covid_urls = []
for i in tqdm(range(len(phishlab['Domain']))):
    time = phishlab['Added'][i]
    date = datetime.strptime(time,'%m/%d/%y')
    phishlab['Added'][i] = date.strftime('%Y-%m-%d')
    #print((int(date[:4]), int(date[5:7]), int(date[8:10])))
    #= 'http://'+ phishlab['Domain'][i]
    #print(phishlab['Domain'][i])
```
