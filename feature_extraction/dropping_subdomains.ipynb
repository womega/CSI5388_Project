```python
def entropy(s):
    p, lns = Counter(s), float(len(s))
    return -sum(count/lns * math.log(count/lns,2) for count in p.values())
```

```python
def avg_entropy(l):
    e = 0
    for i in l:
        e += entropy(i)
    return(e/(len(l)))
```

```python
keywords = ['corona', 'covid' , 'ncov', 'wuhan', 'ncov-19', 
            'virus', 'covid-19', 'covid19', 'sars', 'wuhanvirus',
            'novelvirus']
```

```python
def generate_twists(keys):
    fuzz_doms = []
    fake_doms = []
    # generate twisted domains from the provided keywords
    
    for i in keys:
        i += '.com' # add a random suffix to pass through the fuzz generator
        fuzz = dnstwist.DomainFuzz(i)
        fuzz.generate()
        fuzz_doms += fuzz.domains[1:] # drop the first result in list as it is the original keyword
        
    # extract domain names from the generated list
    
    for i in fuzz_doms:
        _ , domain, _ = extract(i['domain-name'])
        fake_doms.append(domain)
    
    return fake_doms

def check_typo(twists, array):
    typo_doms = []
    for i in array:
        temp_mark = 0
        for j in twists:
            if (len(j) > 3) and (j in i):
                print(j, i)
                temp_mark += 1        
        if temp_mark > 0:
            typo_doms.append(1)
        else:
            typo_doms.append(0)
        
    print("Detected {:.2f}% of domains with typosquatting.".format((typo_doms.count(1)*100/len(typo_doms))))
    return typo_doms
```

```python
fake_doms = generate_twists(keywords)
```

```python
doms = ["na.com", "coronaa.com", "c0vid.com", "coronac.com"]
```

```python
check_typo(fake_doms, doms)
```
